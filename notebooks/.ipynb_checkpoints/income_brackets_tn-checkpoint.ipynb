### Importing Packages and Reading in CSVs

```python
# Importing packages needed for the project
import geopandas as gpd
import pandas as pd
import folium
import seaborn as sns
import folium.plugins as plugins
from folium.plugins import MarkerCluster
from folium.plugins import FastMarkerCluster
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Reading in the geospatial and non-geospatial datasets
tn_pop = pd.read_csv('../data/estimated_pop_TN_counties_2018.csv')
county_income = pd.read_csv('../data/some_TN_county_income.csv')
county_geo = gpd.read_file('../data/county/tncounty.shp')
```

### Exploring the Data

```python
# Looking at the shape of the county income dataset to see how many rows and columns it has
county_income.shape
```

```python
# Looking at the contents of the dataset to see what cleaning has to be done
county_income.head(16)
```

```python
# Checking the tail of the dataframe to see what other parts of the dataset need to be cleaned
county_income.tail()
```

```python
# For my use case, I don't need rows 11, 12, 13, 14, 15, or 16
county_income = county_income.drop([11, 12, 13, 14, 15, 16])
```

```python
# Double checking that the rows got dropped correctly
county_income.head()
```

```python
# Double checking that the rows got dropped correctly
county_income.tail()
```

```python
# Cleaning up the column headers by using a str.replace to get rid of unecessary text
county_income.columns = county_income.columns.str.replace(', Tennessee!!', ' ').str.replace('!!Estimate', '')
```

```python
# Checking to make sure the column headers changed properly
county_income.head()
```

```python
# There were some columns near the end of the dataset I didn't need so I filtered them out
county_income_filtered = county_income.drop(county_income.filter(like = 'Tennessee!!').columns, 1)
```

```python
# Checking to make sure the columns dropped properly
county_income_filtered.head()
```

```python
# Checking the shape after cleaning
county_income_filtered.shape
```

```python
# Looking at the dtypes of the columns to see which I need to change
county_income_filtered.info()
```

### Experimenting with tidying the dataframe

```python
# Transposing the dataset so I have a tidy data layout
county_income_transpose = county_income_filtered.transpose()
```

```python
# I need to reset the index due to the transposed column headers becoming the new index
county_income_transpose.reset_index(inplace = True)
```

```python
# Seeing what other parts of the dataset need to be cleaned
county_income_transpose.head(10)
```

```python
# Checking to make sure the transpose worked properly
county_income_transpose.shape
```

```python
# Looking at the transposed dataset
county_income_transpose.info()
```

```python
# Changing the names of the columns to properly match what I need them to represent
county_income_transpose.columns = ['COUNTY', 'Total Income Reports Filed',
                 'Less Than $10,000', '$10,000 to $14,999',
                 '$15,000 to $24,999', '$25,000 to $34,999',
                '$35,000 to $49,999', '$50,000 to $74,999',
                '$75,000 to $99,999', '$100,000 to $149,999',
                '$150,000 to $199,999', '$200,000 or more']
```

```python
# Dropping the first row which included the original headers I needed for the column names
county_income_transpose = county_income_transpose.drop(county_income_transpose.index[0])
county_income_transpose.reset_index(drop = True)
```

### Creating a County Income Dataframe Only for Households

```python
# Making the dataset smaller by only including household incomes
county_households = county_income_transpose.loc[county_income_transpose['COUNTY'].str.contains('Households')]
county_households = county_households.reset_index(drop = True)
```

```python
# Checking to make sure the dataset got cleaned properly
county_households.head()
```

```python
# Checking the shape of the filtered dataset
county_households.shape
```

```python
# Making the county column be consistent with the other datasets so I can merge them together later
county_households['COUNTY'] = county_households['COUNTY'].str.replace('Households', '')
```

```python
# Checking that the column got properly cleaned
county_households.head()
```

### Started Work on County Geo Shp Dataframe

```python
# Checking the head of the county geo dataframe
county_geo.head()
```

```python
# Checking the info of the county geo dataframe
county_geo.info()
```

```python
# Checking the crs of the county geo dataframe
county_geo.crs
```

```python
# Matching the county column to the county households dataframe so I can merge them later
county_geo = county_geo.rename(columns = {'NAME':'COUNTY', 'KEY':'ZIP'})
county_geo['COUNTY'] = county_geo['COUNTY'] + ' County '
```

```python
# Droppin unecessary columns
county_geo.drop(labels = ['OBJECTID', 'ZIP', 'SHAPE_AREA', 'SHAPE_LEN'], axis = 1, inplace = True)
```

```python
# Checking to make sure the dataframe got cleaned correctly
county_geo.head()
```

### Creating a list to iterate the county dataframe through

```python
# The first step returns a column from the dataframe as a pandas.series
county_list = county_households['COUNTY']
# The second step returns the pandas.series as a class list
county_list = county_list.to_list()
# The third step prints the class type
print(type(county_list))
# The final step prints the list
county_list
```

### Creating Final County Geodataframe

```python
# Using pandas query, I filtered through the county geo dataframe to only include counties that were in the county households dataset
county_geo_final = county_geo.query('COUNTY in @county_list')
county_geo_final.reset_index(drop = True)
```

### Started Work on TN Pop Dataframe

```python
# Checking the head of the tn pop dataset
tn_pop.head()
```

```python
# Dropping unecessary columns and matching the county column to the other datasets so I can merge them later
tn_pop = tn_pop.drop(['Estimates Base', 'Census', '2010', '2011', '2012', '2013', 
                              '2014', '2015', '2016', '2017', '2018'], axis = 1)
tn_pop = tn_pop.rename(columns= {'Counties':'COUNTY', '2019': 'pop'})
```

```python
#tn_pop['COUNTY'] = tn_pop['COUNTY'].str.replace('.','').str.replace(', Tennessee', ' ')
```

```python
# Filtering through the tn pop dataset using pandas query to only return counties in the county households dataset
tn_pop['COUNTY'] = tn_pop['COUNTY'].str.replace('.','').str.replace(', Tennessee', ' ')
tn_pop_filtered = tn_pop.query('COUNTY in @county_list')
tn_pop_filtered.reset_index(drop = True)
```

### Making Final Dataframe

```python
# Trying to merge the three dataframes in one go
county_final = county_geo_final.merge(tn_pop_filtered, on='COUNTY').merge(county_households, on='COUNTY')
```

```python
# Checking to see if the dataframes did merge in one merge statement
county_final.head()
```

```python
# Checking the info on the dataframe
county_final.info()
```

### Experimenting With Plotting Charts

```python
# Experimenting with plotting
county_final.plot(figsize = (8,8));
```

```python
# Creating an area center for the folium map
area_center = [36.4580833, -85.9417256]
```

```python
# Experimenting with folium map
county_points = folium.Map(location = area_center, zoom_start = 7)

folium.GeoJson(county_final).add_to(county_points)

for row_index, row_values in county_final.iterrows():
    center = row_values['geometry'].centroid
    loc = [row_values['geometry'].centroid.y, row_values['geometry'].centroid.x]
    icon = folium.Icon(color = "green", icon = "street-view", prefix = 'fa')
    tool = 'Population: ' + str(row_values['pop']) + '<br/>' + 'Total Income Reports Filed: ' + str(row_values['Total Income Reports Filed'])
        
        
        
    marker = folium.Marker(
            location = loc,
            tooltip = tool,
        icon = icon)
        
    marker.add_to(county_points)

county_points
```
